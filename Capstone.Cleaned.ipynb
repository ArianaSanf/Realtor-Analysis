```python
#Import libraries

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Import data frame

realtor_df= pd.read_excel(r"C:\Users\arire\OneDrive\Desktop\UTD\Data Analytics Bootcamp Fullstack Academy\Capstone\realtor-cleaning.xlsx")
display(realtor_df)
```

```python
#Make zip codes and int value

realtor_df['zip_code'] = realtor_df['zip_code'].astype('Int64')
display(realtor_df)
```

```python
#Check the data types

realtor_df.dtypes
```

```python
#Fill in zip_code missing leading zeros

realtor_df['zip_code'] = realtor_df['zip_code'].apply(lambda x: '{0:0>5}'.format(x))
display (realtor_df)
```

```python
#Find average price by zip code

# Grouping by 'zip_code' to calculate the average price
average_price_by_zip = realtor_df.groupby('zip_code')['price'].mean().reset_index()

# 'average_price_by_zip' is now the go to for av price
display(average_price_by_zip)
```

```python
#Fill in missing values in 'price' column in realtor_df with values from 'average_price_by_zip'

# Calculate average prices by zip code
average_prices = average_price_by_zip.groupby('zip_code')['price'].mean().reset_index()

# Create a mapping between zip codes and the corresponding average prices
zip_avg_price_mapping = dict(zip(average_prices['zip_code'], average_prices['price']))

# Fill in missing 'price' values in the 'realtor' DataFrame
realtor_df['price'] = realtor_df.apply(lambda row: zip_avg_price_mapping.get(row['zip_code'], row['price']), axis=1)

#round by 2 decimal places
realtor_df['price'] = realtor_df['price'].round(2)
```
